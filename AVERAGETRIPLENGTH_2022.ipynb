```python
def label_income (hh):
 if hh['HHFAMINC']==1:
   return '10000'
 if hh['HHFAMINC']==2:
   return '12500'
 if hh['HHFAMINC']==3:
   return '20000'
 if hh['HHFAMINC']==4:
   return '30000'
 if hh['HHFAMINC']==5:
   return '42500'
 if hh['HHFAMINC']==6:
   return '62500'
 if hh['HHFAMINC']==7:
   return '87500'
 if hh['HHFAMINC']==8:
   return '112500'
 if hh['HHFAMINC']==9:
   return '137500'
 if hh['HHFAMINC']==10:
   return '175000'
 if hh['HHFAMINC']==11:
   return '200000'
```

```python
def equivalent_income1 (df):
 if ((df['HHSIZE']==1) & (df['label_income']<= 12500)):
   return 'Poor'
 if ((df['HHSIZE']==2) & (df['label_income']<= 15877)):
   return 'Poor'
 if ((df['HHSIZE']==3) & (df['label_income']<= 19515)):
   return 'Poor'
 if ((df['HHSIZE']==4) & (df['label_income']<= 25094)):
   return 'Poor'
 if ((df['HHSIZE']==5) & (df['label_income']<= 29714)):
   return 'Poor'
 if ((df['HHSIZE']==6) & (df['label_income']<= 33618)):
   return 'Poor'
 if ((df['HHSIZE']==7) & (df['label_income']<= 38173)):
   return 'Poor'
 if ((df['HHSIZE']==8) & (df['label_income']<= 42684)):
   return 'Poor'
 if ((df['HHSIZE']>=9) & (df['label_income']<= 50681)):
   return 'Poor'
 else:
   return 'Not-Poor'
```

```python
def equivalent_income2 (df):
 if ((df['HHSIZE']==1) & (df['label_income']<= 25000)):
   if (df['equivalent_income']=="Poor"):
     return 'Poor'
   else:
     return 'Low-income'
 if ((df['HHSIZE']==2) & (df['label_income']<= 31754)):
   if (df['equivalent_income']=="Poor"):
     return 'Poor'
   else:
     return 'Low-income'
 if ((df['HHSIZE']==3) & (df['label_income']<= 39030)):
   if (df['equivalent_income']=="Poor"):
     return 'Poor'
   else:
     return 'Low-income'
 if ((df['HHSIZE']==4) & (df['label_income']<= 50188)):
   if (df['equivalent_income']=="Poor"):
     return 'Poor'
   else:
     return 'Low-income'
 if ((df['HHSIZE']==5) & (df['label_income']<= 59428)):
   if (df['equivalent_income']=="Poor"):
     return 'Poor'
   else:
     return 'Low-income'
 if ((df['HHSIZE']==6) & (df['label_income']<= 67236)):
   if (df['equivalent_income']=="Poor"):
     return 'Poor'
   else:
     return 'Low-income'
 if ((df['HHSIZE']==7) & (df['label_income']<= 76346)):
   if (df['equivalent_income']=="Poor"):
     return 'Poor'
   else:
     return 'Low-income'
 if ((df['HHSIZE']==8) & (df['label_income']<= 85368)):
   if (df['equivalent_income']=="Poor"):
     return 'Poor'
   else:
     return 'Low-income'
 if ((df['HHSIZE']>=9) & (df['label_income']<= 101362)):
   if (df['equivalent_income']=="Poor"):
     return 'Poor'
   else:
     return 'Low-income'
 else:
   return 'Not-Poor'
```

```python
def CRES(df_1):
   if ((df_1['HHVEHCNT']==0)& ((df_1['equivalent_income']== "Low-income") | (df_1['equivalent_income']== "Poor"))):
     return 'Carless'
   if ((df_1['HHVEHCNT']>=1) & ((df_1['equivalent_income']== "Low-income") | (df_1['equivalent_income']== "Poor"))):
     return 'CRES'
   else:
     return 'Other'
```

```python
def race_name(df):
  if df['R_RACE'] == 1:
    return 'White'
  elif df['R_RACE'] == 2:
    return 'Black or African American'
  elif df['R_RACE'] == 3:
    return 'Asian'
  elif df['R_RACE'] == 4:
    return 'American Indian or Alaska Native'
  elif df['R_RACE'] == 5:
    return 'Native Hawaiian or other Pacific Islander'
  elif df['R_RACE'] == 6:
    return 'Multiracial'
```

```python
def age_range(df):
  if ((df['R_AGE_IMP'] >= 5) & (df['R_AGE_IMP'] <= 15)):
    return '5-15'
  elif ((df['R_AGE_IMP'] >= 16) & (df['R_AGE_IMP'] <= 24)):
    return '16-24'
  elif ((df['R_AGE_IMP'] >= 25) & (df['R_AGE_IMP'] <= 39)):
    return '25-39'
  elif ((df['R_AGE_IMP'] >= 40) & (df['R_AGE_IMP'] <= 64)):
    return '40-64'
  elif ((df['R_AGE_IMP'] >= 65) & (df['R_AGE_IMP'] <= 84)):
    return '65+'
  elif (df['R_AGE_IMP'] >= 85):
    return '85+'
```

```python
def gender_id(df):
  if (df['R_SEX_IMP'] == 1):
    return 'Male'
  if (df['R_SEX_IMP'] == 2):
    return 'Female'
```
