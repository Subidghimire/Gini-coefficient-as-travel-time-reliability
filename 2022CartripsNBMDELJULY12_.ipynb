```python
def label_income (hh):
 if hh['HHFAMINC']==1:
   return '10000'
 if hh['HHFAMINC']==2:
   return '12500'
 if hh['HHFAMINC']==3:
   return '20000'
 if hh['HHFAMINC']==4:
   return '30000'
 if hh['HHFAMINC']==5:
   return '42500'
 if hh['HHFAMINC']==6:
   return '62500'
 if hh['HHFAMINC']==7:
   return '87500'
 if hh['HHFAMINC']==8:
   return '112500'
 if hh['HHFAMINC']==9:
   return '137500'
 if hh['HHFAMINC']==10:
   return '175000'
 if hh['HHFAMINC']==11:
   return '200000'
```

```python
def equivalent_income1 (df):
 if ((df['HHSIZE']==1) & (df['label_income']<= 14800)):
   return 'Poor'
 if ((df['HHSIZE']==2) & (df['label_income']<= 18900)):
   return 'Poor'
 if ((df['HHSIZE']==3) & (df['label_income']<= 23280)):
   return 'Poor'
 if ((df['HHSIZE']==4) & (df['label_income']<= 29950)):
   return 'Poor'
 if ((df['HHSIZE']==5) & (df['label_income']<= 35510)):
   return 'Poor'
 if ((df['HHSIZE']==6) & (df['label_income']<= 40160)):
   return 'Poor'
 if ((df['HHSIZE']==7) & (df['label_income']<= 45690)):
   return 'Poor'
 if ((df['HHSIZE']==8) & (df['label_income']<= 51010)):
   return 'Poor'
 if ((df['HHSIZE']>=9) & (df['label_income']<= 60300)):
   return 'Poor'
 else:
   return 'Not-Poor'
```

```python
def equivalent_income2 (df):
 if ((df['HHSIZE']==1) & (df['label_income']<= 2*14800)):
   if (df['equivalent_income']=="Poor"):
     return 'Poor'
   else:
     return 'Low-income'
 if ((df['HHSIZE']==2) & (df['label_income']<= 2*18900)):
   if (df['equivalent_income']=="Poor"):
     return 'Poor'
   else:
     return 'Low-income'
 if ((df['HHSIZE']==3) & (df['label_income']<= 2*23280)):
   if (df['equivalent_income']=="Poor"):
     return 'Poor'
   else:
     return 'Low-income'
 if ((df['HHSIZE']==4) & (df['label_income']<= 2*29950)):
   if (df['equivalent_income']=="Poor"):
     return 'Poor'
   else:
     return 'Low-income'
 if ((df['HHSIZE']==5) & (df['label_income']<= 2*35510)):
   if (df['equivalent_income']=="Poor"):
     return 'Poor'
   else:
     return 'Low-income'
 if ((df['HHSIZE']==6) & (df['label_income']<= 2*40160)):
   if (df['equivalent_income']=="Poor"):
     return 'Poor'
   else:
     return 'Low-income'
 if ((df['HHSIZE']==7) & (df['label_income']<= 2*45690)):
   if (df['equivalent_income']=="Poor"):
     return 'Poor'
   else:
     return 'Low-income'
 if ((df['HHSIZE']==8) & (df['label_income']<= 2*51010)):
   if (df['equivalent_income']=="Poor"):
     return 'Poor'
   else:
     return 'Low-income'
 if ((df['HHSIZE']>=9) & (df['label_income']<= 2*60300)):
   if (df['equivalent_income']=="Poor"):
     return 'Poor'
   else:
     return 'Low-income'
 else:
   return 'Not-Poor'
```

```python
def CRES(df):
   if ((df['HHVEHCNT']==0)& ((df['equivalent_income']== "Low-income") | (df['equivalent_income']== "Poor"))):
     return 'Carless'
   if ((df['HHVEHCNT']>=1) & ((df['equivalent_income']== "Low-income") | (df['equivalent_income']== "Poor"))):
     return 'CRES'
   else:
     return 'Other'
```

```python
def race_name(df):
  if df['R_RACE'] == 1:
    return 'White'
  elif df['R_RACE'] == 2:
    return 'Black or African American'
  elif df['R_RACE'] == 3:
    return 'Asian'
  elif df['R_RACE'] == 4:
    return 'American Indian or Alaska Native'
  elif df['R_RACE'] == 5:
    return 'Native Hawaiian or other Pacific Islander'
  elif df['R_RACE'] == 6:
    return 'Multiracial'
```

```python
def age_range(df):
  if ((df['R_AGE_IMP'] >= 5) & (df['R_AGE_IMP'] <= 15)):
    return '5-15'
  elif ((df['R_AGE_IMP'] >= 16) & (df['R_AGE_IMP'] <= 24)):
    return '16-24'
  elif ((df['R_AGE_IMP'] >= 25) & (df['R_AGE_IMP'] <= 39)):
    return '25-39'
  elif ((df['R_AGE_IMP'] >= 40) & (df['R_AGE_IMP'] <= 64)):
    return '40-64'
  elif ((df['R_AGE_IMP'] >= 65) & (df['R_AGE_IMP'] <= 84)):
    return '65+'
  elif (df['R_AGE_IMP'] >= 85):
    return '85+'
```

```python
def gender_id(df):
  if (df['R_SEX_IMP'] == 1):
    return 'Male'
  if (df['R_SEX_IMP'] == 2):
    return 'Female'
```
